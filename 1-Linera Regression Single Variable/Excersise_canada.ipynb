```python
import pandas as pd

# Sample DataFrames
data1 = {
    'ID': [1],
    'Name': ['John']
}

data2 = {
    'ID': [2, 3, 4],
    'Age': [25, 30, 22]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Merge the DataFrames based on the common column 'ID'
merged_df = pd.merge(df1, df2, on='ID', how='outer')

# 'how' parameter specifies the type of merge (outer, inner, left, right)
# 'on' parameter specifies the common column for merging

# Display the resulting merged DataFrame
print(merged_df)
```

```python
import pandas as pd

# Sample Parent DataFrame
data_parent = {
    'ID': [1, 2, 3],
    'Name': ['John', 'Alice', 'Bob']
}

# Sample Child DataFrame
data_child = {
    'Age': [25, 30, 22],
    'Country': ['USA', 'Canada', 'UK']
}

df_parent = pd.DataFrame(data_parent)
df_child = pd.DataFrame(data_child)

# Concatenate the DataFrames vertically while preserving all values
merged_df = pd.concat([df_parent, df_child], axis=0, ignore_index=True)

# 'axis' parameter specifies the axis for concatenation (0 for rows, 1 for columns)
# 'ignore_index' resets the index of the resulting DataFrame

# Display the resulting concatenated DataFrame
print(merged_df)
```

```python
import pandas as pd

# Sample Parent DataFrame
data_parent = {
    'ID': [1, 2, 3],
    'Name': ['John', 'Alice', 'Bob']
}

# Sample Child DataFrame
data_child = {
    'Age': [25, 30, 22],
    'Country': ['USA', 'Canada', 'UK']
}

df_parent = pd.DataFrame(data_parent)
df_child = pd.DataFrame(data_child)

# Concatenate the DataFrames vertically to add child data below the parent data
merged_df = pd.concat([df_parent, df_child], axis=0, ignore_index=True)

# 'axis' parameter specifies the axis for concatenation (0 for rows, 1 for columns)
# 'ignore_index' resets the index of the resulting DataFrame

# Display the resulting DataFrame with child data added below parent data
print(merged_df)
```
