```python
data = tt_df[['PassengerId'	,'Survived'	,'Pclass','Age'	,'SibSp',	'Parch',	'Fare','Sex']]
data
```

```python
dummy_df = pd.get_dummies(data.Sex )
dummy_df
```

```python
data=data.drop(['Sex'], axis='columns')
data
```

```python
dummy_df.isnull().sum()
```

```python
fdata = pd.concat([data,dummy_df],axis='columns')
fdata.head()
```

```python
X = fdata[['PassengerId','Pclass','Age',	'SibSp'	,'Parch'	,'Fare',	'female',	'male']]
```

```python
Y = fdata['Survived']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
trainX,testX,trainY,testY = train_test_split(X,Y , test_size=0.2)
```

```python
len(testY)
```

```python
model = LogisticRegression()
```

```python
model.fit(trainX,trainY)
```

```python
model.score(testX,testY)
```
