```python
import pandas as pd
```

```python
car_price = pd.read_csv('carprices.csv')
car_price
```

```python
dummies = pd.get_dummies(car_price["Car Model"])
dummies
```

```python
car_price = pd.concat([car_price , dummies] , axis='columns')
car_price
```

```python
car_price = car_price.drop(["Car Model"] , axis='columns')
car_price
```

```python
car_price = car_price.drop(["Mercedez Benz C class"] , axis='columns')
car_price
```

```python
X = car_price.drop('Sell Price($)', axis='columns')
X
```

```python
Y = car_price["Sell Price($)"]
```

```python
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression()
model
```

```python
model.fit(X,Y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.predict([[86000,7,0,1]])
```

```python
model.score(X,Y)
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
```

```python
car_price_sk = pd.read_csv('carprices.csv')
car_price_sk
```

```python
lE_car_price = car_price_sk
```

```python
lE_car_price["Car Model"] = LE.fit_transform(lE_car_price["Car Model"])
lE_car_price
```

```python
X = lE_car_price[['Car Model' , 'Mileage' , 'Age(yrs)']].values
X
```

```python
Y = lE_car_price["Sell Price($)"].values
Y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer([("Car Model" , OneHotEncoder() ,[0] )],remainder='passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X= X[:,1:]
X
```

```python
model.fit(X,Y)
```

```python
model.predict([[0,1,45000,4]])
```

```python
model.predict([[0,0,86000,7]])
```
